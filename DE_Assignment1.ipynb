```python
#replacing value by Mean

# imputing with SimpleImputor from the sklearn library
from sklearn.impute import SimpleImputer

print("Data of Social_Life before replacing")
print(data['Social_Life'])


print("")
print("Mean of Social_Life")
Social_Life_mean_value = round(data["Social_Life"].mean(), 1)
print(Social_Life_mean_value)


# define the imputer
imr = SimpleImputer(missing_values=np.nan, strategy='mean') # or median
data[['Social_Life']] = imr.fit_transform(data[['Social_Life']])
print(data['Social_Life'])
```
