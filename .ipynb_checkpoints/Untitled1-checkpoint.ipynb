```python
data1=pd.read_csv('cars.csv')
col_Drop = ['MPG']
data1.drop(col_Drop,axis=1,inplace=True)
print(data1)
```

```python
data.fillna(value=0,method=None,axis=None,inplace=False,limit=None,downcast=None)
```

```python
mean_value=round(data['Cylinders'].mean(),2)
print("Replace By Mean Value")
data['Cylinders'].fillna(mean_value,inplace=True)
print(data)


median_value=round(data['Cylinders'].median(),2)
print("Replace By Median Value")
data['Cylinders'].fillna(median_value,inplace=True)
print(data)
```

```python
data1.bfill(inplace=True)
print(data1)
```

```python
data1.ffill(inplace=True)
print(data1)
```
