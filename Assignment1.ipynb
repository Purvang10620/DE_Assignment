```python
import pandas as pd
import numpy as np
data1=pd.read_csv('cars.csv')

col_Drop = ['MPG']
data1.drop(col_Drop, axis=1, inplace=True)

print(data1)
```

```python
#replace missing value by 0 in dataset

data1.fillna(value=0, method=None, axis=None, inplace=False, limit=None, downcast=None)
```

```python
# replacing missing value by mean value

Weight_mean_value = round(data1['Weight'].mean(), 2)
data1['Weight'].fillna(Weight_mean_value, inplace=True)

print("Relpacing By Mean Value")
print(data1)

# replacing missing value by Median value

Weight_median_value = round(data1['Weight'].median(), 2)
data1['Weight'].fillna(Weight_median_value, inplace=True)

print("Relpacing By Median Value")
print(data1)

# replacing missing value by Mode value

Weight_mode_value = round(data1['Weight'].mode(), 2)
data1['Weight'].fillna(Weight_mode_value, inplace=True)

print("Relpacing By Mode Value")
print(data1)
```

```python
import pandas as pd
import numpy as np
# imputing with SimpleImputor from the sklearn library
from sklearn.impute import SimpleImputer

# define the imputer
imr = SimpleImputer(missing_values=np.nan, strategy='mean') # or median
data[['Cylinders','Displacement']] = imr.fit_transform(data[['Cylinders','Displacement']])
print(data[0:20])
```
